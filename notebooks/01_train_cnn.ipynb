```python
import os
from ultralytics import YOLO
from pathlib import Path
import torch
```

```python
SELECTED_DATASET = 'garbage_dataset_100_reduction'
YOLO_VERSION = 'yolov8n.pt'
MODEL_NAME = 'eco_sorter_v2'
```

```python
BASE_DIR = Path('..').resolve()
DATA_YAML_PATH = BASE_DIR / 'data' / 'dataset' / SELECTED_DATASET / 'data.yaml'
DATASET_DIR = BASE_DIR / 'data' / 'dataset' / SELECTED_DATASET 

print(f"Chemin vers data.yaml : {DATA_YAML_PATH}")
```

```python
# Count files
train_imgs = len(list((DATASET_DIR / 'train' / 'images').glob('*')))
train_lbls = len(list((DATASET_DIR / 'train' / 'labels').glob('*')))
val_imgs = len(list((DATASET_DIR / 'valid' / 'images').glob('*')))
val_lbls = len(list((DATASET_DIR / 'valid' / 'labels').glob('*')))

print(f"\n✓ Files in dataset:")
print(f"  Training: {train_imgs} images, {train_lbls} labels")
print(f"  Validation: {val_imgs} images, {val_lbls} labels")
```

```python
model = YOLO(str(YOLO_VERSION))

training_args = {
    'data': str(DATA_YAML_PATH), # Path to data.yaml
    'epochs': 30,                # Number of epochs (adjust based on dataset size)
    'imgsz': 640,                # Image size
    'batch': 8,                 # Batch size (reduce if GPU memory is limited)
    'patience': 10,              # Early stopping patience
    'save': True,                # Save checkpoints
    'device': 0 if torch.cuda.is_available() else 'cpu',  # Use GPU if available
    'project': os.path.join(BASE_DIR, 'models_training_runs'), # Project directory
    'name': str(MODEL_NAME),     # Experiment name
    'exist_ok': False,            # Overwrite existing project
    'pretrained': True,          # Use pretrained weights
    'optimizer': 'AdamW',        # Optimizer (SGD, Adam, AdamW)
    'lr0': 0.001,                # Initial learning rate
    'weight_decay': 0.0005,      # Weight decay
    'warmup_epochs': 3,          # Warmup epochs
    'augment': True,             # Use data augmentation
    'verbose': True,             # Verbose output
}

# 3. Lancer l'entraînement
# project=... permet de sauvegarder les résultats proprement dans un dossier 'models' à la racine
model.train(
    **training_args
)
```
